```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Relation between tips & total bill

sns.relplot(x='tip',y='total_bill',data=tips)
```

```python
# Sepration with respect to smoker
sns.relplot(x='tip',y='total_bill',data=tips,hue='smoker')
```

```python
# Sepration with respect to day & style is smoker
sns.relplot(x='tip',y='total_bill',data=tips,hue='day',style='smoker')
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,hue='day',style='smoker',size='size')
```

```python
# Create visulation as per tips size

sns.relplot(x='tip',y='total_bill',data=tips,size='tip')
```

```python
# Load Iris DataSet
iris = sns.load_dataset('iris')
```

```python
# Inside relpot you can use many type of graph like scatter, line
# But you can plot scatterplot without relplot

sns.scatterplot(x = 'sepal_length',y = 'sepal_width',data = iris )
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Understand that customer dense is high on Sat & Sunday
sns.catplot(x = 'day',y='tip',data=tips)
```

```python
# Who has given highest tips
sns.catplot(x = 'smoker',y='tip',data=tips)
```

```python
# Box Plot to get an intiution of outlier (IQR)
sns.catplot(x='day',y='tip',kind='box',data=tips)
```

```python
# Who has given a more steps Male or Female

sns.catplot(x='day',y='tip',kind='box',data=tips,hue='sex')
```

```python
# Generate random data with normal distribution
x = np.random.randn(1000)
```

```python
# We can see distribution of dataset.
sns.distplot(x)
```

```python
# See the data distribution along with relation between two data
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(iris)
```

```python
# draw Line 
sns.lmplot(x = 'total_bill',y='tip',data = tips)
```

```python
# regression plot & lm plot is almost same. Shaded region is nothing but weight & biased that are used while fitting a best fit line
sns.regplot(x='total_bill', y = 'tip' , data = tips)
```

```python
# Daywise plot
sns.lmplot(x = 'total_bill',y='tip',data = tips,col='day')
```
